```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('provinces_updated.xls')
df

df_sorted = df.sort_values(by='Population', ascending=True)
df_sorted

plt.figure(figsize=(13, 5))
plt.bar(df_sorted.index + 1, df_sorted['Population'], color='skyblue')  
plt.ylabel('Population (in thousands)')
plt.xlabel('Province/City Number')
plt.title('Population of Provinces and Cities (Sorted in Ascending Order)')


plt.xticks(ticks=df_sorted.index + 1, labels=(df_sorted.index + 1))
plt.tight_layout()
plt.show()


province_mapping = {i + 1: name for i, name in enumerate(df_sorted['Name'])}
province_mapping
```

```python
df = pd.read_excel('region_areas.xls')
df

plt.figure(figsize=(8, 8))
plt.pie(df['Area Proportion (%)'], labels=df["Region"], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Area Proportion by Region')
plt.tight_layout()
plt.show()
```

```python
population_density = df['PopulationDensity']

plt.figure(figsize=[10,10]) 
plt.xticks(range(50, int(population_density.max()) + 1000, 230))
plt.hist(population_density, bins=20, color='skyblue', edgecolor='black') 

plt.title('Frequency of Population Density in VietNam Provinces & Cities', fontsize=14)
plt.xlabel('Population Density (People / km2)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
cyclone = pd.read_csv('cyclones.csv')
cyclone

plt.figure(figsize=(12, 6))
plt.plot(cyclone['date'], cyclone['cyclone_events'], marker='o', color='#759CBC', linestyle='-', linewidth=2, markersize=5)

plt.title('Cyclones events in VietNam', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of events', fontsize=14)


plt.tight_layout()
plt.show()
```
